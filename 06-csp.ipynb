#### 練習問題 6.1.7: 魔法列



```python
from cspsat import *
def magicSequence(n, x=Bool("x")):
    for i in range(n):
        yield ["eqK", [ x(i,j) for j in range(n) ], 1]
        for j in range(n):
            yield ["equ", x(i,j), ["eqK", [ x(k,i) for k in range(n) ], j]]

x = Bool("x")
for n in range(1, 16):
    print(f"# n = {n}")
    for sol in solutionsCSP(magicSequence(n), num=0):
        print([ j for i in range(n) for j in range(n) if sol[x(i,j)] ])
```

#### 練習問題 6.1.8: 合成数と素数



```python
from cspsat import *
def composite(m, n=Var("n")):
    yield ["int", n, 2, m]
    cs = [ ["and", ["<", k, n], ["==", ["mod", n, k], 0]] for k in range(2, m) ]
    yield ["or", *cs]

def prime(m, n=Var("n")):
    yield ["int", n, 2, m]
    for k in range(2, m+1):
        yield ["imp", ["<", k, n], [">", ["mod", n, k], 0]]
```

```python
m = 50
solveCSP(composite(m), num=0)
```

```python
m = 50
solveCSP(prime(m), num=0)
```
